```python
from ultralytics import YOLO
import glob
from PIL import Image
```

```python
#Training block
```

```python
# Load a model
model = YOLO('yolov8x.yaml')  # build a new model from YAML

# Train the model
results = model.train(data='data.yaml', epochs=75, imgsz=224, batch=40, patience= 300,verbose=True, hsv_h=0, hsv_s=0, hsv_v=0)
```

```python
#Inference blocks
```

```python
model = YOLO(r'last.pt')  # load a partially trained model

lst=glob.glob(r"C:\Users\ritvi\AWCEBC\data\test\Auto-WCEBleedGen Challenge Test Dataset\Test Dataset 2/*.png") 
results = model(lst, stream=True)  # return a generator of Results objects
```

```python
i=0
# Show the results
for r in results:
    im_array = r.plot()  # plot a BGR numpy array of predictions
    im = Image.fromarray(im_array[..., ::-1])  # RGB PIL image
    im.save(r"C:\Users\ritvi\AWCEBC\data\test\Auto-WCEBleedGen Challenge Test Dataset\Test Dataset 2\preds/"+str(i)+".jpeg")  #change root path to where you want to save file
    i+=1
```
